```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample data (replace with your actual data)
# First plot: Change in position (z) in cm with error in measurement on the x-axis
z = np.array([0.8,1.6,2.3,3.1,3.5])  # Position in cm
F = np.array([0.00981,0.01962,0.02943,0.03924,0.04905])  # Force in Newtons
z_error = 0.1  # Error in position (z) measurement

# Second plot: Delta z on y-axis and I (measured current) on x-axis
delta_z = np.array([-.2,-.5,-.7,-.9,-1.1,-1.3,-1.5])  # Change in position (delta z)
I = np.array([.2,.403,.601,.801,1.001,1.203,1.4])  # Measured current in Amperes


# Fit for the first plot (F vs. z)
slope_F_z, intercept_F_z, r_value, p_value, std_err_F_z = stats.linregress(z, F)

# Fit for the second plot (delta_z vs. I)
slope_delta_z_I, intercept_delta_z_I, r_value, p_value, std_err_delta_z_I = stats.linregress(I, delta_z)

# Create the first plot: F vs. z with error bars
plt.figure(figsize=(8, 5))  # Adjusted figure size
plt.errorbar(z, F, xerr=z_error, fmt='o', label='Data Points', capsize=5, elinewidth=2)
plt.plot(z, slope_F_z * z + intercept_F_z, 'r-', label=f'Fit: F = ({slope_F_z:.2f})z + ({intercept_F_z:.2f})')
plt.xlabel('Position (z) [cm]')
plt.ylabel('Force (F) [N]')
plt.title('Force vs. Position with Best Fit Line')
plt.legend()
plt.grid(True)

# Annotating the slope and error on the plot
plt.text(3.5, 2.0, f'Slope: {slope_F_z:.2f}  {std_err_F_z:.2f}', fontsize=12, color='blue')
plt.text(3.5, 1.8, f'Intercept: {intercept_F_z:.2f}', fontsize=12, color='blue')

# Adjust layout to remove empty space
plt.subplots_adjust(left=0.12, right=0.88, top=0.88, bottom=0.12)

# Show the plot
plt.show()

# Create the second plot: delta_z vs. I with error bars
plt.figure(figsize=(8, 5))  # Adjusted figure size
plt.errorbar(I, delta_z, xerr=0.01, fmt='o', label='Data Points', capsize=5, elinewidth=2)
plt.plot(I, slope_delta_z_I * I + intercept_delta_z_I, 'g-', label=f'Fit: delta_z = ({slope_delta_z_I:.2f})I + ({intercept_delta_z_I:.2f})')
plt.xlabel('Current (I) [A]')
plt.ylabel('Change in Position (delta_z) [cm]')
plt.title('Change in Position vs. Current with Best Fit Line')
plt.legend()
plt.grid(True)

# Annotating the slope and error on the plot
plt.text(0.55, 0.2, f'Slope: {slope_delta_z_I:.2f}  {std_err_delta_z_I:.2f}', fontsize=12, color='green')
plt.text(0.55, 0.15, f'Intercept: {intercept_delta_z_I:.2f}', fontsize=12, color='green')

# Adjust layout to remove empty space
plt.subplots_adjust(left=0.12, right=0.88, top=0.88, bottom=0.12)

# Show the plot
plt.show()
```
