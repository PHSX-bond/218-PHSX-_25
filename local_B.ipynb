```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

theta_100 = np.array([20.8, 21.5, 30.0, 30.0, 41, 39, 50, 49.5, 61, 60])
I_100 = np.array([7.16, 7.57, 11.50, 11.45, 16.76, 16.69, 25.2, 25.1, 38.4, 38.4])

theta_200 = np.abs(np.array([21, -20, 32, -31, 39, -38, 48, -47, 60, -58]))
I_200 = np.abs(np.array([3.64, -3.59, 6.45, -6.44, 7.80, -7.79, 11.02, -11.02, 18.04, -17.94]))

theta_300 = np.abs(np.array([18, -18, 30, -29, 38, -38, 49, -48, 59, -58]))
I_300 = np.abs(np.array([1.94, -2.34, 3.76, -3.90, 5.01, -5.46, 7.75, -8.44, 11.83, -11.81]))

tan_theta_100 = np.tan(np.radians(theta_100))
tan_theta_200 = np.tan(np.radians(theta_200))
tan_theta_300 = np.tan(np.radians(theta_300))

error = 0.5

fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].errorbar(tan_theta_100, I_100, yerr=error, fmt='bo', label="Data", capsize=5)
slope_100, intercept_100, r_value_100, p_value_100, std_err_100 = stats.linregress(tan_theta_100, I_100)
axs[0].plot(tan_theta_100, slope_100 * tan_theta_100 + intercept_100, 'r-', label=f'Fit Line (slope = {slope_100:.2f})')
axs[0].set_title("I vs. tan() for N=100")
axs[0].set_xlabel("tan()")
axs[0].set_ylabel("I")
axs[0].legend()
axs[0].text(0.1, 45, f"Slope = {slope_100:.2f}\nR = {r_value_100**2:.2f}", fontsize=12, verticalalignment='bottom')

axs[1].errorbar(tan_theta_200, I_200, yerr=error, fmt='go', label="Data", capsize=5)
slope_200, intercept_200, r_value_200, p_value_200, std_err_200 = stats.linregress(tan_theta_200, I_200)
axs[1].plot(tan_theta_200, slope_200 * tan_theta_200 + intercept_200, 'r-', label=f'Fit Line (slope = {slope_200:.2f})')
axs[1].set_title("I vs. tan() for N=200")
axs[1].set_xlabel("tan()")
axs[1].set_ylabel("I")
axs[1].legend()
axs[1].text(0.1, 35, f"Slope = {slope_200:.2f}\nR = {r_value_200**2:.2f}", fontsize=12)

axs[2].errorbar(tan_theta_300, I_300, yerr=error, fmt='ro', label="Data", capsize=5)
slope_300, intercept_300, r_value_300, p_value_300, std_err_300 = stats.linregress(tan_theta_300, I_300)
axs[2].plot(tan_theta_300, slope_300 * tan_theta_300 + intercept_300, 'r-', label=f'Fit Line (slope = {slope_300:.2f})')
axs[2].set_title("I vs. tan() for N=300")
axs[2].set_xlabel("tan()")
axs[2].set_ylabel("I")
axs[2].legend()
axs[2].text(0.1, 35, f"Slope = {slope_300:.2f}\nR = {r_value_300**2:.2f}", fontsize=12)

plt.tight_layout()
plt.show()


```
